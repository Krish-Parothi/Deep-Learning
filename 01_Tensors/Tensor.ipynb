### Creating Tensor

```python
a = torch.empty(2,3)
# empty function create space in memory, i.e allocates memory, and do not assign value on that space. only the values which are over there will get shown up.
# empty(2,3) will create 2 dimensional vector of size 2x3
```

```python
a
```

```python
type(a) # Tensor Object
```

```python
# using Zeroes
b = torch.zeros(2,3)
# This makes a tensor of shape of your choice, and all values are assigned as 0.
```

```python
b
```

```python
# using Ones
c = torch.ones(2,3)
# This makes a tensor of shape of your choice, and all values are assigned as 1.
```

```python
c
```

```python
# using rand
d = torch.rand(2,3)
# This function makes tensor of given shape, and values are randomly assigned, with value between 0 to 1.
```

```python
d
```

```python
# Now if we want that same random value that we generated using rand, then we use seed.

# Means we want same output every time.

torch.manual_seed(100)
torch.rand(2,3)
```

```python
torch.manual_seed(100)
torch.rand(2,3)
```

```python
# custom tensor using tensor
torch.tensor([[1,2,3],[4,5,6]])
```

```python
# other ways

# arange
print("using arange =>", torch.arange(0,10,2)) # 0 to 10 is range and 2 is for jumping ahead of 2 numbers.
print("using arange =>", torch.arange(0,10)) # 0 to 10 is range and 2 is for jumping ahead of 2 numbers.

# using linspace
print("using linspace =>", torch.linspace(0,10,10))


# using eye
print("using eye =>", torch.eye(5)) # eye is Identity Matrix

# using full
print("using full =>", torch.full((3,3),5)) # This Means we want 3x3 matrix with each item as 5.
```

#### Tensor Shapes

```python
x = torch.tensor([[1,2,3],[4,5,6]])
x
```

```python
x.shape
```

```python
torch.empty_like(x)
# If humko koi existing tensor ka shape copy krna hai then use empty_like() Function. This creates new tensor with same shape as x and it has different values.
```

```python
torch.zeros_like(x) # This creates new tensors but all values are zeros
```

```python
torch.ones_like(x) # This creates new tensors but all values are ones.
```

```python
torch.rand_like(x) # This code will not work because rand will generate float values between 0 and 1, and above we made the tensor x as all integers. so here code will not work.
```

### Mathematical Operations

#### 1. Scalar Operations

```python
x = torch.rand(2,2)
x
```

```python
# Addition
x + 2
# Substraction
x - 2
# Multiplication
x * 3
# Division
x / 3
# int Division
(x*100)//3
# Mod
((x*100)//3)%2
# Power
x**2
```

### 2. Element Wise Operation

```python
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
```

```python
# add 
a+b
```

```python
# subtraction
a-b
```

```python
# multiplication 
a*b
```

```python
# Division
a/b
```

```python
# power
a**b
```

```python
# mod 
a % b
```

```python
c = torch.tensor([1,-2,3,-4])
```

```python
d = torch.tensor([1.9, 2.3, 5.6,4.5])
```

```python
# round (round off)
torch.round(d)
```

```python
# ceil
torch.ceil(d)
```

```python
# floor
torch.floor(d)
```

```python
# clamp
torch.clamp(d, min=2, max=3)
```
