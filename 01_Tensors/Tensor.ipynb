### Creating Tensor

```python
a = torch.empty(2,3)
# empty function create space in memory, i.e allocates memory, and do not assign value on that space. only the values which are over there will get shown up.
# empty(2,3) will create 2 dimensional vector of size 2x3
```

```python
a
```

```python
type(a) # Tensor Object
```

```python
# using Zeroes
b = torch.zeros(2,3)
# This makes a tensor of shape of your choice, and all values are assigned as 0.
```

```python
b
```

```python
# using Ones
c = torch.ones(2,3)
# This makes a tensor of shape of your choice, and all values are assigned as 1.
```

```python
c
```

```python
# using rand
d = torch.rand(2,3)
# This function makes tensor of given shape, and values are randomly assigned, with value between 0 to 1.
```

```python
d
```

```python
# Now if we want that same random value that we generated using rand, then we use seed.

# Means we want same output every time.

torch.manual_seed(100)
torch.rand(2,3)
```

```python
torch.manual_seed(100)
torch.rand(2,3)
```

```python
# custom tensor using tensor
torch.tensor([[1,2,3],[4,5,6]])
```

```python
# other ways

# arange
print("using arange =>", torch.arange(0,10,2)) # 0 to 10 is range and 2 is for jumping ahead of 2 numbers.
print("using arange =>", torch.arange(0,10)) # 0 to 10 is range and 2 is for jumping ahead of 2 numbers.

# using linspace
print("using linspace =>", torch.linspace(0,10,10))


# using eye
print("using eye =>", torch.eye(5)) # eye is Identity Matrix

# using full
print("using full =>", torch.full((3,3),5)) # This Means we want 3x3 matrix with each item as 5.
```

### 3. Reduction Operation

```python
# sum along columns
torch.sum(e, dim=0) # dimension = 0 for column
```

```python
# sum along rows
torch.sum(e, dim=1) # dimension = 1 for row
```

```python
# mean
torch.mean(e) # mean only work if torch type is float
```

```python
# mean along column
torch.mean(e, dim=0) # dimension = 0 for mean along column
```

```python
# median
torch.median(e) 
```

```python
torch.max(e)
```

```python
torch.min(e)
```

```python
torch.prod(e) # saare items ka product
```

#### 4. Matrix Operations

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)

print(f)
print(g)
```

```python
# matrix Multiplication
torch.matmul(f,g)
```

```python
# dot products
vector1 = torch.tensor([1,2])
vector2 = torch.tensor([3,4])

torch.dot(vector1, vector2)
```

```python
# transpose
f
```

```python
torch.transpose(f, 0, 1) # columns ko rows bnan do
```

```python
h = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
h
```

```python
torch.det(h) # determinant works on square matrix
```

```python
# inverse
torch.inverse(h) # inverse works on square matrix
```

```python
# greater than
i > j
```

```python
# less than
i < j
```

```python
# equal to
i == j
```

```python
# not equal to
i != j
```

```python
# greater than equal to
i>=j
```

```python
# less than equal to
i<=j
```
